```python
import unittest
```

```python
def extended_euclidean(a, b):
    # Series of remainders
    c = a
    d = b
    
    s, ss, t, tt = 1, 0, 0, 1
    
    while d != 0:
        q = c // d
        
        # c, d = d, c % d
        c, d = d, c - q*d
        
        s, ss = ss, s - q * ss
        t, tt = tt, t - q * tt
        
    return c, s, t
```

```python
class TestExtendedEuclidean(unittest.TestCase):
    def test_extended_euclidean(self):
        gcd, m, n = extended_euclidean(33, 30)
        self.assertEqual(gcd, 3)
        self.assertEqual(m, 1)
        self.assertEqual(n, -1)
        
        gcd, m, n = extended_euclidean(150, 241)
        self.assertEqual(gcd, 1)
        self.assertEqual(m, -98)
        self.assertEqual(n, 61)
        
        gcd, m, n = extended_euclidean(221, 188)
        self.assertEqual(gcd, 1)
        self.assertEqual(m, 57)
        self.assertEqual(n, -67)
        
        gcd, m, n = extended_euclidean(372, 403)
        self.assertEqual(gcd, 31)
        self.assertEqual(m, -1)
        self.assertEqual(n, 1)
        
        gcd, m, n = extended_euclidean(380, 218)
        self.assertEqual(gcd, 2)
        self.assertEqual(m, 35)
        self.assertEqual(n, -61)
        
    def test_probabilistic(self):
        for i in range(1000):
            a = randint(0, 2^10)
            b = randint(0, 2^10)
            
            g, m, n = extended_euclidean(a, b)
            self.assertEqual(gcd(a, b), g)
            self.assertEqual(m * a + n * b, g)
                    
result = unittest.main(argv=[''], exit = False)
```
